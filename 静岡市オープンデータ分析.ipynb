```r
#CSVファイルの読み込み
csv_data <- read.csv("shizuoka_R.csv")
csv_data
nrow(csv_data)
```

```r
#グラフで用いるラベルを作成
labels=c("0-4歳","5-9歳","10-14歳","15-19歳","20-24歳","25-29歳","30-34歳","35-39歳","40-44歳","45-49歳","50-54歳","55-59歳","60-64歳","65-69歳","70-74歳","75-79歳","80-84歳","85歳以上")
```

```r
#日本語の対応
par(family = "HiraKakuProN-W3")
```

```r
index <- 832 #indexは行番号を指定
pie(as.integer(c(csv_data[index,5:22])),labels = labels, 
    main = paste(csv_data[index,2],csv_data[index,3]),
    clockwise=TRUE, col = rainbow(18))
```

```r
barplot(as.integer(c(csv_data[index,5:22])),legend = labels,
        main = paste(csv_data[index,2],csv_data[index,3]),
        names.arg = labels, col = rainbow(18))
```

```r
barplot(t(csv_data[1:100,5:22]), legend = labels,
        col = rainbow(18), names.arg = csv_data[1:100,3]) 
#10件の町別年齢層
```

```r
des_data <- csv_data[csv_data$区=="葵区",]
des_data <- des_data[order(des_data$総数, decreasing = TRUE),]
des_data
```

```r
barplot(t(des_data[1:100,5:22]), legend = labels,
        col = rainbow(18), names.arg = des_data[1:100,3]) 
```

```r
columnList <- c("総数", "X70.74歳","X75.79歳","X80.84歳","X85歳以上")
scatter <- csv_data[csv_data$区=="葵区",columnList,]
scatter_y <- scatter[,2]+scatter[,3]+scatter[,4]+scatter[,5]
plot(scatter[,1], scatter_y, xlab = "Population", ylab = "70 Over")
result.lm <- lm(scatter_y ~ scatter[,1])
summary(result.lm)
abline(result.lm)
mtext(substitute(paste(R^2," = ", text),list(text = round(summary(result.lm)$adj.r.squared,digits = 5))),line=-1,col=2,adj=0)
```
